```python
import numpy as np
import pandas as pd
import os
from os.path import isfile , join
```

```python
resource_path = './Resources'
only_files = [join(resource_path, f) for f in os.listdir(resource_path) if isfile(join(resource_path, f)) 
              and f.find('.csv') != -1]
dfs = []
for f in only_files:
    df = pd.read_csv(f)
    dfs.append(df)
storm_events = pd.concat(dfs)

storm_events.to_csv('StormEvents_details.csv', sep = ',')
storm_events
```

```python
storm_events["YEAR"] = storm_events["BEGIN_YEARMONTH"].apply(lambda x: int(str(x)[0:4]))
```

```python
hurricane = storm_events[storm_events.apply(
    lambda x: x["EVENT_TYPE"].find('Hurricane')!= -1 or x["EVENT_TYPE"].find('Tropical Storm')!= -1, axis=1)]
hurricane
```

```python
hurricane.to_csv('hurricane_tropical_storms.csv')
```
